```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
y=df.iloc[:,12:] # prediction of cardio using other parameters
y
```

```python
from sklearn.model_selection import train_test_split  
#split 80% data into training data and rest 20% as testing data
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8)
```

# Using Different Models to predict Cardio

```python
svm_score=model1.score(xtest,ytest)
print(f"The accuracy of SVM model is : {svm_score}")
```

# Hyperparameter Tuning for SVM (Support Vector Machine)

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
param_grid_svm = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'gamma': ['scale', 'auto'],
    'degree': [3, 4, 5]
}
grid_search_svm = GridSearchCV(SVC(), param_grid_svm, cv=5, n_jobs=-1, verbose=1)
```

```python
grid_search_svm.fit(xtrain, ytrain)
```

```python

print("Best parameters for SVM:", grid_search_svm.best_params_)
best_svm_model = grid_search_svm.best_estimator_
```

```python

svm_score = best_svm_model.score(xtest, ytest)
print(f"The tuned accuracy of SVM model is : {svm_score}")
```
